## Knn model without temperature (Low bleu scores ~19.5)

```python
# Load knns data
data = pickle.load(open("/private/home/urvashik/gen-knn/outputs/ccnc_k8.pkl", "rb"))
print("Number of examples = %d" % len(data))
print(data[0].keys())
dct = dictionary.Dictionary.load("/private/home/urvashik/gen-knn/data-bin/wmt18.de-en/dict.en.txt")
dstore_vals = np.memmap('/checkpoint/urvashik/dstores/cc_nc14.de-en/endstore_vals.npy', dtype=np.int, mode='r', shape=(67680775, 1))
```

```python
first_ex = data[0]
print('id =', first_ex['id'])
print('src =', dct.string(first_ex['src']))
print('tgt =', dct.string(first_ex['tgt']))
print(first_ex['hypo'][0].keys())
hypo = first_ex['hypo'][0]
print(hypo['tokens'])
print(hypo['vals'])
```

```python
samples = np.load("/checkpoint/urvashik/dstores/cc_nc14.de-en/dstore.samples.npy")
cumsum = np.cumsum(samples[:,1], axis=0)
cumsum -= 1 # because the indexing in the dstore starts at 0
def get_sample_id(val):
    index = cumsum.searchsorted(val)
    return samples[index], cumsum[index-1], cumsum[index]
```

```python
def print_knns_for_ex(ex):
    hypo = ex['hypo'][0]
    print('Print Knns:')
    for step in range(len(hypo['tokens'])):
        print('Source = %s' % dct.string(ex['src']))
        print('Reference target = %s' % dct.string(ex['tgt']))
        print('Model output = %s' % dct.string(hypo['tokens']))
        print('Model output token %d = %s\n' % (step, dct.string(hypo['tokens'][step:step+1])))
        knns = hypo['knns'][step]
        vals = hypo['vals'][step]
        probs = hypo['probs'][step]
        sample, cumsum_starts, cumsum_currs = get_sample_id(knns)
        sample_id = sample[:,0]
        sample_len = sample[:,1]
        #print('knn index = %d, sample_id = %d, sample_len = %d' % (knn, sample_id, sample_len))
        print('Retrieved values = %s' % dct.string(vals))
        for k in range(knns.shape[0]):
            outstr = dct.string(dstore_vals[cumsum_starts[k]:cumsum_currs[k]].reshape(-1))
            print('sample for k=%d=%s, sample_id=%d, lprob=%.4f\nREF: %s\nSRC: %s\n' % (k, dct.string([vals[k]]), sample_id[k], probs[k], outstr, deen[outstr]))

        print('==================\n==================')
```
